```python
import gzip
import io
import pandas as pd
import urllib.request

%matplotlib inline

import matplotlib # plotting
import matplotlib.pyplot as plt

import sqlite3

%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python

ramphs = pd.read_csv("http://sebastianheath.com/roman-amphitheaters/roman-amphitheaters.csv")
chronogrps = pd.read_csv("http://sebastianheath.com/roman-amphitheaters/chronogrps.csv")
```

```python
ramphs_sql = sqlite3.connect(":memory:")
```

```python
ramphs.to_sql('ramphs',ramphs_sql,if_exists = "replace")
chronogrps.to_sql('chronogrps',ramphs_sql,if_exists = "replace")
```

```python
response = urllib.request.urlopen("http://atlantides.org/downloads/pleiades/dumps/pleiades-places-latest.csv.gz")
pleiades = pd.read_csv(io.BytesIO(gzip.decompress(response.read())))
```

```python
pleiades.to_sql('pleiades',ramphs_sql,if_exists = "replace")
```

```python
cur = ramphs_sql.cursor()
cur.execute("UPDATE pleiades SET path = 'https://pleiades.stoa.org'||path")
cur.close()
```

```python
pd.read_sql_query("SELECT path FROM pleiades LIMIT 5", ramphs_sql)
```

```python
# 1) Adapt the above SELECT query to do something like:
# "Find all amphitheaters in Italy from sites that Pleiades thinks were founded before 323 BC. Order by earliest date of the amphitheater." 
# Hint: Use the pleiades.minDate column
# Then make some variations on that theme

pd.read_sql_query('''SELECT  ramphs.id, ramphs.modcountry, pleiades.minDate
          FROM ramphs, pleiades
          WHERE ramphs.pleiades = pleiades.path and ramphs.modcountry = "Italy" and pleiades.minDate < -322
          ORDER BY pleiades.minDate;''', ramphs_sql)
```

```python
#all the amphitheaters from France in the hellenistic-republican,roman,late-antique (RLM) period with a capacity greater than 5000
pd.read_sql_query('''SELECT  ramphs.id, ramphs.modcountry, ramphs.capacity, pleiades.timePeriods
          FROM ramphs, pleiades
          WHERE ramphs.pleiades = pleiades.path 
              and ramphs.modcountry = "France" 
              and pleiades.timePeriods = "RLM" 
              and ramphs.capacity > 4999
          ORDER BY ramphs.capacity;''', ramphs_sql)
```

```python
#add chronogrps.start information in to pull from three tables
pd.read_sql_query('''SELECT  ramphs.id, ramphs.modcountry, ramphs.capacity, pleiades.timePeriods, chronogrps.start
          FROM ramphs, pleiades, chronogrps
          WHERE ramphs.pleiades = pleiades.path 
              and ramphs.chronogrp = chronogrps.chronogrp
              and ramphs.modcountry = "France" 
              and pleiades.timePeriods = "RLM" 
              and ramphs.capacity > 4999
          ORDER BY ramphs.capacity;''', ramphs_sql)
```

```python
#amphitheaters built in Itally after 70 BCE
pd.read_sql_query('''SELECT  ramphs.id, ramphs.modcountry, chronogrps.start
          FROM ramphs, chronogrps
          WHERE ramphs.chronogrp = chronogrps.chronogrp
              and ramphs.modcountry = "Italy" 
              and chronogrps.start > -71 
          ORDER BY chronogrps.start;''', ramphs_sql)
```

```python
#all amphitheaters from Italy after 70 BCE with a extmajor greater than 80 and an extminor greater than 60
pd.read_sql_query('''SELECT  ramphs.id, ramphs.modcountry, ramphs.extmajor, ramphs.extminor, pleiades.minDate
          FROM ramphs, pleiades
          WHERE (ramphs.pleiades = pleiades.path)
          and (ramphs.modcountry = "Italy")
          and (ramphs.extmajor > 79)
          and (ramphs.extminor > 59)
          and (pleiades.minDate > -71)
          ORDER BY pleiades.minDate;''', ramphs_sql)
```
